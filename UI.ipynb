Demo1: Microphone or uploaded audio

```python
from transformers import pipeline
import gradio as gr

pipe = pipeline(model="Antares28/whisper-tiny-zh")  # change to "your-username/the-name-you-picked"

def transcribe(audio):
    text = pipe(audio)["text"]
    return text

iface = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(type="filepath"),
    outputs="text",
    title="Whisper Tiny Chinese",
    description="Realtime demo for Chinese speech recognition using a fine-tuned Whisper tiny model.",
)

iface.launch()
```

Demo2: URL

```python
from transformers import pipeline
import gradio as gr
import yt_dlp
from moviepy.editor import AudioFileClip

pipe = pipeline(model="Antares28/whisper-tiny-zh")

def transcribe(video_url):
    try:
      # use youtube-dl to download video
      ydl_opts = {
          'format': 'bestaudio/best',
          'outtmpl': 'downloaded_video.%(ext)s',
          'noplaylist': True,
      }
      with yt_dlp.YoutubeDL(ydl_opts) as ydl:
          ydl.download([video_url])

      audio_clip = AudioFileClip("downloaded_video.m4a").subclip(0,20)
      audio_clip.write_audiofile("audio.wav")

      # use Whisper model to transcribe
      text = pipe("audio.wav")["text"]

      return text
    except Exception as e:
      print("Excetion:", e)
      return "An error occurred during processing, please check if the URL is correct or see more error messages."


iface = gr.Interface(
    fn=transcribe,
    inputs=gr.Textbox(placeholder="Enter Video URL Here..."),
    outputs="text",
    title="Whisper Tiny Chinese",
    description="Realtime demo for Chinese speech recognition using a fine-tuned Whisper tiny model.",
)

iface.launch()
```
